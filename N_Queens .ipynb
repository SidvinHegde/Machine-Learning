```python
#  N_Queens Problem



import random
def rowAttacks(board_array):
    attacks=0
    for i in range(len(board_array)):
        for j in range(i+1,len(board_array)):
            if(board_array[i]==board_array[j]):
                attacks+=1
    return attacks

def increasingDiagonalAttacks(board_array):
    attacks=0
    for i in range(len(board_array)):
        for j in range(i+1,len(board_array)):
            if(board_array[i]+j-i == board_array[j]):
                attacks+=1
    return attacks

def decreasingDiagonalAttacks(board_array):
    attacks=0
    for i in range(len(board_array)):
        for j in range(i+1,len(board_array)):
            if(board_array[i]-(j-i) == board_array[j]):
                attacks+=1
    return attacks

def score(board_array):
    total_score = rowAttacks(board_array)+increasingDiagonalAttacks(board_array)+\
    decreasingDiagonalAttacks(board_array)
    return total_score



def neibhours(board_array):
    neibhours_array = []
    for i in range(len(board_array)):
        for j in range(len(board_array)):
            if(board_array[i]!=j):
                new_board_array_position = list(board_array)
                new_board_array_position[i]=j
                neibhours_array.append(new_board_array_position)
    return neibhours_array

def bestNeibhour(board_array,current_score):
    neibhours_array = neibhours(board_array)    
    scores = list(map(lambda x: score(x), neibhours_array))
    best_score = min(scores)
    if(best_score<= current_score):
        indexes = []
        for index,value in enumerate(scores):
            if(value==best_score):
                indexes.append(index)
        best_index = random.choice(indexes)
        return neibhours_array[best_index]
    
def printBoard(board_array):
    board_squares = []
    for i in board_array:
        row=[]
        for j in board_array:
            row.append('_')
        board_squares.append(row)
        
    
    for i in range(len(board_array)):        
        board_squares[board_array[i]][i]= 'Q'    
    for i in board_squares:
        print()
        for j in i:
            print(j,end='\t')
    print('\n  attacks=',score(board_array))

def NQueensProblem(n):
    board_position =[]
    for i in range(n):
        board_position.append(random.randint(0,n-1))
    
    current_score = score(board_position)
    while(current_score>0):
        printBoard(board_position)
        best_next_state = bestNeibhour(board_position,current_score)
        if(not best_next_state):
            print("hit local minima exiting!")
            break        
        board_position= best_next_state
        current_score= score(board_position)
    if(current_score==0):
        printBoard(board_position)
        
NQueensProblem(4)
    
    
    
    
```
